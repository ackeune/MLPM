### Notation

$\mathbf{X}$ is the $M \times T$ data matrix, containing $M$ measurements at $T$ time steps.

$\mathbf{S}$ is the $S \times T$ source matrix, containing $S$ source signal values at $T$ time steps. We will assume $S = M$.

$\mathbf{A}$ is the mixing matrix. We have $\mathbf{X} = \mathbf{A S}$.

$\mathbf{W}$ is the matrix we aim to learn. It is the inverse of $\mathbf{A}$, up to indeterminacies (scaling and permutation of sources).

$\phi$ is an elementwise non-linearity or activation function, typically applied to elements of $\mathbf{W X}$.

In the following assignments, you can make use of the signal generators listed below.



```python
%pylab inline

# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
    return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp

def sine_wave(x, period=0.2, amp=1.0, phase=0.):
    return np.sin((x / period - phase) * 2 * np.pi) * amp

def square_wave(x, period=0.2, amp=1.0, phase=0.):
    return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) * 2 * amp

def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp

def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix if shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X):
    """
    Plot the signals contained in the rows of X.
    """
    figure()
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
```

### 1.3 Implicit priors (20 points)
As explained in MacKay's book, an activation function $\phi$ used in the ICA learning algorithm corresponds to a prior distribution over sources. Specifically, $\phi(a) = \frac{d}{da} \ln p(a)$. For each of the following activation functions, derive the source distribution they correspond to.
$$\phi_0(a) = -\tanh(a)$$
$$\phi_1(a) = -a + \tanh(a)$$
$$\phi_2(a) = -a^3$$
$$\phi_3(a) = -\frac{6a}{a^2 + 5}$$

The normalizing constant is not required, so an answer of the form $p(a) \propto \verb+[answer]+$ is ok.

Plot the activation functions and the corresponding prior distributions. Compare the shape of the priors to the histogram you plotted in the last question.

$$\newcommand{\defeq}{\stackrel{\textrm{def}}{=}}$$
$$\newcommand{\abs}{\textrm{abs}}$$
#### Answer:
The feature functions are defined as:
$$ \phi(a) \defeq \frac{d}{da} \ln p(a)$$
Therefore the implicit prior $p(a)$ is proportional to the exponent of the antiderivative of $\phi(a)$ (as the normalization constants in $p(a)$, which are not dependent on $a$, will disappear when taking the derivative w.r.t. $a$ of the logarithm of $p(a)$):

$$ \ln p(a) \propto \int \phi(a) da \Rightarrow p(a) \propto e^{\int \phi(a) da} $$

For the $\phi$ functions in the assignment we then have: 

1. $\phi_0(a) = -\tanh(a) \rightarrow \ln p(a) \propto -\int \tanh(a)da$. Therefore: 

    $$\tanh(a) \defeq \frac{\sinh(a)}{\cosh(a)} = \frac{\frac{1}{2}(e^a-e^{-a})}{\frac{1}{2}(e^a+e^{-a})} = \frac{e^a-e^{-a}}{e^a+e^{-a}}$$

    We can use the fact that the numerator is the derivative of the denominator (and vice versa), which can easily be seen from the fact that $\frac{d}{da}(e^a+e^{-a})=\frac{d}{da}e^a+\frac{d}{da}e^{-a}=e^a-e^{-a}$, by definition of derivatives of exponents with base $e$ and the chain rule of differentiation. Let $u=e^a+e^{-a}$. Then $du = e^a-e^{-a}da$. Thus:
    \begin{align*}
        \int \tanh(a)da &= \int \frac{du}{u}\\
		&= \int \frac{1}{u} du\\
		&= \ln\abs(u)\\
		&= \ln \abs(e^a + e^{-a})\\	
		&= \ln (e^a + e^{-a})
	\end{align*}
	as $\forall (a)[e^a + e^{-a} > 0]$. Thus $-\int \tanh(a)da=-\ln(e^a+e^{-a})$. 
		
	Therefore $\phi(a)$ corresponds to an underlying distribution 
	\begin{align*}
		& &p(a)&\propto e^{-\int\tanh(a)da}\\
		&\Rightarrow &p(a)&\propto e^{-\ln(e^a+e^{-a})}\\
		&\Rightarrow &p(a)&\propto\frac{1}{e^a+e^{-a}}\\
        &\Rightarrow &p(a)&\propto\frac{2}{e^a+e^{-a}} = \textrm{sech}(a)
	\end{align*}
    The last step can be taken because the normalization constant is not yet known.
		
2.	$\phi_1(a) = -a + \tanh(a) \Rightarrow \ln p(a) \propto \int -a + \tanh(a)da$

	$-\int [a + \tanh(a)] da = -\int a da + \int \tanh(a) da = -\frac{1}{2}a^2 + \ln(e^a + e^{-a})$.
	Therefore, $\phi_1(a)$ implies an underlying distribution 
    \begin{align*}
    & 			&p(a)&\propto e^{-\int [a+\tanh(a)]da}\\
    &\Rightarrow	&p(a)&\propto e^{-\frac{1}{2}a^2 + \ln(e^a+e^{-a})}\\
    &\Rightarrow	&p(a)&\propto e^{-\frac{1}{2}a^2}e^{\ln(e^a+e^{-a})}\\
    &\Rightarrow	&p(a)&\propto e^{-\frac{1}{2}a^2}(e^a + e^{-a})\\
    &\Rightarrow	&p(a)&\propto e^{-\frac{1}{2}a^2+a}+e^{-\frac{1}{2}a^2-a}
    \end{align*}

3.	$\phi_2(a) = -a^3 \Rightarrow \ln p(a) \propto -\int a^3da$.

    $-\int a^3 da = -\frac{1}{4}a^4$. Therefore, $\phi_0(a)$ implies an underlying distribution $p(a) \propto e^{-\frac{1}{4}a^4}$.
		  
4.  $\phi_3(a) = -\frac{6a}{a^2+5}=-6\frac{a}{a^2+5} \Rightarrow \ln p(a) \propto -6\int\frac{a}{a^2+5}da$.
    
    From http://integral-table.com/downloads/integral-table.pdf we have that the antiderivative of a function of the form $f(x) = \frac{x}{x^2+a}$ is $\frac{1}{2}\ln \abs(x^2 + a)$.
    Therefore: $-6\int\frac{a}{a^2+5}da = -6(\frac{1}{2}\ln \abs(a^2 + 5))$.
    Thus, $\phi_3(a)$ corresponds to a probability distribution $p(a) \propto e^{-3\ln \abs(a^2 + 5)}$, which simplifies to $e^{-3\ln \abs(a^2 + 5)}=(e^{\ln \abs(a^2+5)})^{-3} = \abs(a^2+5)^{-3}=\frac{1}{(a^2+5)^3}$ as $\forall(a)[x^2+5>0]$.

Now compute the covariance matrix of the sources, the measurements and the whitened measurements. You can visualize a covariance matrix using the line of code below. Are the signals independent after whitening?

```python
%pylab inline
#C = np.eye(5)  # Dummy matrix; compute covariance here

def whiten(X):
    X = numpy.matrix(X)

    #making the data have a zero mean
    X = X - numpy.mean(X)
    #calculating the covariance
    A = numpy.dot(X.T,X)
    #calculating the eigenvalues and eigenvectors
    d,V = numpy.linalg.eigh(A)

    epsilon = 0.1    
    D = numpy.diag(1./numpy.sqrt(d+epsilon))
        
    #whitening matrix
    W = numpy.dot(numpy.dot(V,D),V.T)
    
    #whitening the data
    X = numpy.dot(X,W)
    return X
    
    
Xwhitened = whiten(X)

Xwhitened = np.squeeze(np.asarray(Xwhitened))

C = numpy.cov(Xwhitened)
ax = imshow(C, cmap='gray', interpolation='nearest')
```

$$
\verb+Kurt+[X] = \frac{\mu_4}{\sigma^4} - 3 = \frac{\operatorname{E}[(X-{\mu})^4]}{(\operatorname{E}[(X-{\mu})^2])^2} - 3
$$
Here, $\mu_4$ is known as the fourth moment about the mean, and $\sigma$ is the standard deviation.
The '-3' term is introduced so that a Gaussian random variable has 0 excess kurtosis.
